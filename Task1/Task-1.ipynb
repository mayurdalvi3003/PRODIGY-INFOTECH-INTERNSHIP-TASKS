# 1) Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic.csv")
```

```python
# on x axis i have taken the Age distribution and on Y axis i have taken Fare distribution 
# i am trying to plot a bar chart using matplotlib 
plt.bar(df["Age"], df["Fare"])
plt.title("Visualization of Age and Fare")
plt.xlabel("Age distribution")
plt.ylabel("Fare distribution")
plt.show()
```

# Bar chart visualization of Categotical data using pandas `plot()`function

```python
df["Sex"].value_counts().plot(kind ="bar" , xlabel = "Sex" ,ylabel = "Count_Of_Sex")
```
